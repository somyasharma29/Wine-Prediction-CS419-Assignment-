```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('train.csv')
df.head()
```

```python
df1= pd.read_csv('test.csv')
df1.head()
```

```python
df['type'].unique()
```

```python
df.columns
```

```python
dummies =  pd.get_dummies(df['type'])
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged.head()
```

```python
final= merged.drop(['red','type'], axis=1)
final.head()
```

```python
X= final.drop(['quality'],axis=1)
X.head()
```

```python
y= final.quality
y.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
y2= model.predict(X)
y3= []
for i in range(len(y2)):
    y3.append(round(y2[i]))
y3
```

```python
model.score(X,y3)
```

```python
dummies1 =  pd.get_dummies(df1['type'])
dummies1
```

```python
merged1 = pd.concat([df1,dummies1],axis='columns')
merged1.head()
```

```python
X1= merged1.drop(['red','type'], axis=1)
X1.head()
```

```python
y1= model.predict(X1)
y4= []
for i in range(len(y1)):
    y4.append(round(y1[i]))
y4
```
